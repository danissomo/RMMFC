```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
import skfmm

# Load image
im = imageio.v2.imread("rVHKd.png")
ima = np.array(im)


# Differentiate the inside / outside region
phi = np.int64(np.any(ima[:, :, :3], axis = 2))
print(phi.max(), phi.min())
plt.imshow(phi)
plt.colorbar()
plt.show()
# The array will go from - 1 to 0. Add 0.5(arbitrary) so there 's a 0 contour.
phi = np.where(phi, 1, -1)

# Show phi
plt.imshow(phi)
plt.xticks([])
plt.yticks([])
plt.colorbar()
plt.show()

# Compute signed distance
# dx = cell(pixel) size
sd = skfmm.distance(phi, dx = 1)

# Plot results
plt.imshow(sd)
plt.colorbar()
plt.show()
```
